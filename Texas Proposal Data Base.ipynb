```python
texas_df = pd.read_excel("Texas 2017 Quotes.xlsx")
```

```python
print(texas_df)
```

```python
#  Select only the main columns of the data frame
clean_df_tx = texas_df[['bid_no_calc','BID_YEAR','C_city','C_country','C_cust_name','C_state','C_state','C_zip','contact_true_email','contact_true_first','Customer_ID','intNotes','IX_CUST_ID_SEQ','Market','PROJECT NAME','Total_prop_v','user_est_total','valid']]
clean_df_tx
```

```python
# cleaning the names
rename_df = clean_df_tx.rename(columns={'bid_no_calc':' Proposal #',\
                      'BID_YEAR':'Year',\
                      'C_city':'City',\
                      'C_cust_name':'Customer Name',\
                      'C_state':'State',\
                      'C_zip':'Zip Code',\
                      'contact_true_email':'Email',\
                      'contact_true_first':'Contact Firt Name',\
                      'Customer_ID':'Customer ID',\
                      'intNotes':'Notes',\
                      'Market':'Market',\
                      'PROJECT NAME':'Project Name',\
                      'Total_prop_v':'Proposal Value',\
                      'user_est_total':'Estimated Value',\
                      'valid':'Status'})


rename_df
```

```python
# Sorting the Data Frame 
organized_df= rename_df[[' Proposal #','Market','Project Name','Customer ID','Customer Name','Proposal Value','Estimated Value','Status','Notes','City','Zip Code','Contact Firt Name','Email','Year']]

organized_df

```

```python
# setting up a column "market" as index 
indexNames = organized_df[ organized_df['Market'] == "Water/Wastewater"].index

print(indexNames)
```

```python
# Delete water waste water rows
organized_df.drop(indexNames , inplace=True)
organized_df
```

```python
# Drop Duplicates in Estimated Value
indexDuplicates = organized_df[ organized_df['Status'] == "Duplicate"].index
print(indexDuplicates)
```

```python
organized_df.drop(indexDuplicates , inplace=True)
organized_df
```

```python
# Drop Na values 
organized_df.dropna(axis=0,how='all',thresh=None, inplace =True)
organized_df
```

```python
# Reset Index 
organized_df.reset_index(drop=True, inplace=True)
organized_df


```

```python
# export the data frame to excel
with pd.ExcelWriter('Data_Frame_Tx.xlsx') as writer:  
    organized_df.to_excel(writer, sheet_name='Proposals_17')
```
