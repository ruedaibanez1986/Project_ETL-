```python
# Importing dictionaries 
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

## Store xlsx into Data Frame

```python
# Reading the excel file 

excel_file = 'Resources/Texas 2017 Quotes.xlsx'

texas_df = pd.read_excel(excel_file)
print(texas_df)
```

## Creating a New Data Frame with the selected Columns and Transforming the Names of Columns

```python
#  populating a new df with the columns we need from the original data base - texas-df
clean_df_tx = texas_df[['bid_no_calc','BID_YEAR',\
                        'C_city','C_country',\
                        'C_cust_name',\
                        'C_state',\
                        'C_state',\
                        'C_zip',\
                        'contact_true_email',\
                        'contact_true_first',\
                        'Customer_ID',\
                        'intNotes',\
                        'IX_CUST_ID_SEQ',\
                        'Market',\
                        'PROJECT NAME',\
                        'Total_prop_v',\
                        'user_est_total',\
                        'valid']].copy()
clean_df_tx
```

```python
# Renaming the df with columns name that make more sense  - new data frame name
rename_df = clean_df_tx.rename(columns={'bid_no_calc':'Proposal #',\
                      'BID_YEAR':'Year',\
                      'C_city':'City',\
                      'C_cust_name':'Customer Name',\
                      'C_state':'State',\
                      'C_zip':'Zip Code',\
                      'contact_true_email':'Email',\
                      'contact_true_first':'Contact Firt Name',\
                      'Customer_ID':'Customer ID',\
                      'intNotes':'Notes',\
                      'Market':'Market',\
                      'PROJECT NAME':'Project Name',\
                      'Total_prop_v':'Proposal Value',\
                      'user_est_total':'Estimated Value',\
                      'valid':'Status'})


rename_df
```

```python
# Organizing the df by sorting the columns with better logic and importances 
organized_df= rename_df[['Customer ID','Proposal #','Market','Project Name','Customer Name','Proposal Value','Estimated Value','Status','Notes','City','Zip Code','Contact Firt Name','Email','Year']].copy()
organized_df

```

## Eliminating the Water Waste Water Market and dropping the duplicates Proposals 

```python
# setting up a column "market" as index 
indexNames = organized_df[ organized_df['Market'] == "Water/Wastewater"].index

print(indexNames)
```

```python
# Delete water waste water rows withing the data frame 
organized_df.drop(indexNames , inplace=True)

organized_df
```

```python
# Drop Duplicates in Estimated Value
indexDuplicates = organized_df[ organized_df['Status'] == "Duplicate"].index
print(indexDuplicates)
```

```python
organized_df.drop(indexDuplicates , inplace=True)
organized_df
```

## Eliminating the NA withing the DF

```python
# Drop Na values 
organized_df.dropna(axis=0,how='all',thresh=None, inplace =True)
organized_df
```

## Resetting the Index know the total amount of Data 

```python
# Reset Index 
organized_df.reset_index(drop=True, inplace=True)
print(organized_df


```

## Exporting the Clean and Organized Data Frame 

```python
# export the data frame to csv fole

organized_df.to_csv("Output_files/texas_df_2017.csv",index=False)
```
